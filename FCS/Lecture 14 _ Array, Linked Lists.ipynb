## Array

 - List의 하부구조
 - 여러 element를 그룹화하여 저장하고자 할 때, 순차적으로 넘버링 된 메모리에 element를 저장함.(neighboring rule)
  - 한 개의 element가 저장된 메모리로 접근함으로써 모든 elment에 접근이 가능한 장점. 
    - 가령, L[i]로 해당 element에 매우 간편하게 접근이 가능. (0부터 N-1까지)
  - 단, 처음 선언한 메모리에 따라 그 길이가 결정됨. (fixed integer length (N))
    - 동적 메모리가 할당되는 List와 다른 점.(Array resizing을 통해 array 하부구조로 List를 구현한 것)


## Array Resizing
 - 이슈: N개 items 저장을 하는데, 메모리가 지나치게 낭비되거나, 메모리가 지나치게 부족하거나.
 - 해결 방법: 더 큰 array를 만들어서 모든 items를 copy해버리기
   - 가령, L.append(1) 하면, Len(L)+1인 메모리를 찾아서 새 array를 만든 다음, L의 items를 모두 복붙하고, 마지막 자리에 1을 할당하는 식임.
   - 메모리를 두 배 이상 쓰는 방법이므로 매우 비싼 방법.
 - Python의 개선 방법: 현재 len(list)에 비례해서 resize.
   - 처음 0개일 때 하나 넣을 때는 4로 늘리고,
   - 4개가 다 차면 8로 늘리고
   - 8개 다 차면 16으로 늘리고 .... 
   - 그나마 메모리 낭비가 덜하긴 하지만, append, remove 시에 resize에 따르는 부담이 있는 건 변하지 않음


## Linked Lists
 - 각 Node가 link되어 있는 List 구조. 
 - 각 Node는 하나의 정수만 value로 가지고 있음.
 - 각 Node에 접근하려면, 첫 Node부터 차례차례로 링크를 타고 접근해야 함. 
   - 장점: 각 element가 굳이 순차적으로 연결된(neighboring) 메모리에 있지 않아도 됨. 마구 흩어져 있어도 링크를 통해 모든 메모리에 접근 가능하니까.
   - 즉, 동적 메모리 할당을 실현한 것. resizing 전혀 필요 없음.
   - 단점: 해당 element가 어디있는지 모름... 특정 element를 찾으려면 모든 list를 한번씩은 방문해야 함.
   - 캐싱(방문 없이 저장된 값을 출럭하기), sentinel 방법으로 개선 가능

```python
class LinkedNode():
  def __init__(self, x):
    self.val = x
    self.next = None
```

```python
a = LinkedNode(5)
b = LinkedNode(7)
a.next = b
```

## Single Linked Lists(SLL)
 - Methods 만들기
  - addFirst()
     - 해당 SLL 객체에 first라는 변수에, 첫 번째 노드를 만들어 할당. 
     - 만약 기존에 노드가 있었다면, 기존 노드를 뒤로 보내고, 새 노드를 첫 번째 노드로 변경.
  - getFirst()
    - 해당 SLL 객체의 첫 번째 노드의 value를 출력

```python
class SLList:
  def __init__(self) -> None:
    self.first = None

  def addFirst(self, x: int) -> None:
    newFirst = LinkedNode(x) # 새로운 노드 만듦
    newFirst.next = self.first # 기존 노드는 새로운 노드의 뒤로 연결
    self.first = newFirst # 새로운 노드를 first에 할당하여 첫 번째 노드로 만듦 
  
  def getFirst(self) -> None:
    if self.first: # first가 있으면
      return self.first.val # first에 할당된 첫 번째 노드의 value 출력
    return None # first가 없으면 None 출력
```

 - 추가 Methods 만들기
   - getSize()
     - 해당 SLL의 길이 출력
     - 첫 노드부터 끝 노드까지 타고가면서 사이즈를 카운트
       - 이슈: Time Complex : O(N) 너무 오래 걸림
       - 해결: first 변수처럼, size 변수도 애초에 SLL 객체에 만들어 버리기 (메모리를 더 쓰고, 시간을 줄이기) 

```python
class SLList:
  def __init__(self) -> None:
    self.first = None

  def addFirst(self, x: int) -> None:
    newFirst = LinkedNode(x)
    newFirst.next = self.first 
    self.first = newFirst 
  
  def getFirst(self) -> None:
    if self.first: 
      return self.first.val 
    return None 

  def getSize(self) -> int:
    curNode = self.first
    size = 0
    while curNode: # curNode가 유효한 값이면 아래 계속 반복
      size +=1
      curNode = curNode.next
    return size
```

   - append()
     - 해당 SLL의 맨 끝에 노드 추가

```python
class SLList:
  def __init__(self) -> None:
    self.first = None
    self.size = 0 
  def addFirst(self, x: int) -> None:
    newFirst = LinkedNode(x)
    newFirst.next = self.first 
    self.first = newFirst
    self.size += 1
  
  def getFirst(self) -> None:
    if self.first: 
      return self.first.val 
    return None 

  def getSize(self) -> int:
    return self.size
  
  def append(self, x:int) -> None:
    self.size += 1
    curNode = self.first
    while curNode.next: # curNode가 None이 되기 전까지
      curNode = curNode.next # 링크를 계속 타고 이동. 반복 끝나면 curNode.next == None 인 상태.
    curNode.next = LinkedNode(x) # 이때 curNode.next에 새로운 노드 생성하여 할당
```

```python
L = SLList()
L.addFirst(5)
L.append(10)
```

## Linked Lists with Sentinel Node

 - 특별한 경우를 고려하여 조건을 추가하고자 한다면, 코딩에서 오류 낼 가능성이 더 높아짐
 - 대안: first 변수 대신, Sentinel이라는 변수를 만들어, 여기에 Dummy 노드를 할당. (첫 번째 노드는 Dummy 노드의 next에 할당)
   - 즉, sentinel.next 가 진짜 첫 번째 노드.
   - 장점: append method를 정의할 때, 특별한 경우를 굳이 고려하지 않아도 됨

```python
class SLList:
  def __init__(self) -> None:
    self.sentinel = LinkedNode(-99999) # self.first 대신 self.sentinel 변수 생성 및 더미 노드 할당
    self.size = 0 

  def addFirst(self, x: int) -> None:
    newFirst = LinkedNode(x)
    newFirst.next = self.sentinel.next # 모든 first 대신, sentinel.next를 작성
    self.sentinel.next = newFirst
    self.size += 1
  
  def getFirst(self) -> None:
    if self.sentinel.next: 
      return self.sentinel.next.val 
    return None 

  def getSize(self) -> int:
    return self.size
  
  def append(self, x:int) -> None:
    self.size += 1
    curNode = self.sentinel # sentinel에도 (더미) 노드가 있으므로, 설령 첫 번째 노드가 없다 하더라도 append를 통해 첫 번째 노드를 만들 수 있게됨
    while curNode.next: 
      curNode = curNode.next 
    curNode.next = LinkedNode(x) 
```

```python
L = SLList()
L.append(10)
```

```python
L.getSize()
```

```python
L.addFirst(0)
L.getFirst()
```

```python
L.getSize()
```
