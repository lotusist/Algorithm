 - List의 elements들을 switch 하는 방법: x, y = y, x 
 - 빈 List는 index를 가지지 않는다.
 - if문 조건을 통해 반복 작업이 이루어지는 index를 통제할 수 있어야 한다. 이해 못하겠으면 0부터 range 끝까지 한 번 써보면서 패턴을 파악해야 한다. 
 - deepcopy와 shallow copy의 차이: B = A[:]는 shallow copy로서, 값을 복사하긴 하지만 인스턴스가 메모리에 새로 생성되는 것이 아니다. 즉, 값 자체를 복사하는 게 아니라, 주소값을 복사하여 B가 같은 메모리를 가리키게 하는 것이다. 그래서 B를 수정하면 원본인 A도 수정되어버린다. 

```python
from typing import List
import copy
def flipAndInvertImage(A: List[List[int]]) -> List[List[int]]:
  row = len(A)
  col = len(A[0])
  for i in range(row):
    for j in range(col):
      if j < col-1-j:
        A[i][j], A[i][col-1-j] = A[i][col-1-j], A[i][j]
      else:
        break
  
  for i in range(row):
    for j in range(col):
      A[i][j] = 1 - A[i][j]
      
  return A
```

```python
from typing import List
import copy
def flipAndInvertImage(A: List[List[int]]) -> List[List[int]]:
  B = copy.deepcopy(A)
  # B = A[:]
  row = len(A)
  col = len(A[0])
  for i in range(row):
   for j in range(col):
     B[i][j] = A[i][col-1-j]
  for i in range(row):
    for j in range(col):
      B[i][j] = 1 - B[i][j]
  return B
```

```python
A = [[1,1,0,0], [1,0,0,1],[0,1,1,1],[1,0,1,0]]
B = [[1,1,0],[1,0,1],[0,0,0]]
```

```python
flipAndInvertImage(A)
```

```python
flipAndInvertImage(B)
```
