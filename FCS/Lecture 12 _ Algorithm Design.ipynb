## Two Sum

 - 정수 array, 그리고 정수 target이 주어짐
 - 정수 array의 두 개 element를 더했을 때, target이 되는 유일한 pair가 존재함
 - 해당 pair의 index를 출력
 - 단, Time Complexity: ~ $N^2$ , Space Complexity: ~1

```python
import random
L = list(range(20000))
random.shuffle(L)
```

```python
def twoSum(L: list, target: int)->list:
  for i in range(len(L)):
    for j in range(len(L)):
      if L[i] + L[j] == target:
        return [i,j]
```

```python
import time
t_start = time.perf_counter()
print(twoSum(L,1))
t_end = time.perf_counter()
print(t_end-t_start)
```

```python
def twoSum(L: list, target: int)->list:
  for i in range(len(L)):
    for j in range(i+1, len(L)):
      if L[i] + L[j] == target:
        return [i,j]
```

```python
t_start = time.perf_counter()
print(twoSum(L,1))
t_end = time.perf_counter()
print(t_end-t_start)
```

```python
def twoSum(L: list, target: int)->list:
  myDict = {} # value: index
  for i in range(len(L)):
    myDict[L[i]] = i
  for i in range(len(L)):
    complement = target - L[i]
    if complement in myDict:
      return [i, myDict[complement]]
```

```python
t_start = time.perf_counter()
print(twoSum(L,1))
t_end = time.perf_counter()
print(t_end-t_start)
```

```python
def twoSum(L: list, target: int)->list:
  myDict = {} # value: index
  for i in range(len(L)):
    if L[i] in myDict:
      return [myDict[L[i]], i]
    complement = target - L[i]
    myDict[complement] = i
```

```python
t_start = time.perf_counter()
print(twoSum(L,1))
t_end = time.perf_counter()
print(t_end-t_start)
```

  - Two Pointer Algorithm
   - 조건: Sorted List

```python
def twoSumSorted(L: list, target: int)->list:
  i, j = 0, len(L)-1 # 2 pointers
  while L[i] +L[j] != target:
    if L[i] + L[j] < target:
      i = i+1
    else:
      j = j+1
  return [i,j]
```
