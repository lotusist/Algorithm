## Types of Programming Cost

1. Execution Cost
 - Time Complexity
 - Memory Complexity

2. Programming Cost
 - Development time
 - Readability, Modifiability, Maintainability

### Time Complexity
 - Time Module (import time, per_counter())
 - Cannot use the computer while counting the time... 
 - Results depend on lots of factors (compiler, data, machine,...)

#### Input size N
 - Proportional to input size N
 - Machine independent
 - Can evaluate the algorithm's scalability
 - Really tedious to compute
 - Not an actual time consumed

#### Big O
  - With a very large N (Let's assume Big Data environment)
  - Kind of Asymptotic Analysis
  - Order of growth is important: the number of exponential is influential to the shape of the function.
   - 1) consider the worst case
   - 2) focus on the highest order of growth
   - 3) remove the lower order terms
   - 4) remove the constants
