## Merge sort

 - 단계
  - 1) 전체 리스트를 2개 서브 리스트로 쪼개기
  - 2) left sub와 right sub를 각각 sort
  - 3) 2개 sub를 제대로 된 방식으로 sort하여 합침

 - 구현: recursive 구조 활용. 각 sub에 merge sort를 시행.
    - 1) sub의 길이가 1이 될 때까지 쪼개기
    - 2) sub의 길이가 1이 됨 (base case)
    - 3) 이제 sub를 합쳐가기 시작함.(merge) 두 개 sub의 가장 왼쪽에 있는 두 element들을 비교하여, 둘 중 작은 것이 merged list의 가장 왼쪽으로 가도록 함
    - 4) 가장 작은 element가 제외된 두 개 sub의 다시 가장 왼쪽에 있는 두 element들을 비교하여, 둘 중 작은 것이 merged list의 가장 왼쪽으로 가도록 함. 반복.   


```python
def mergeSort(L:list) -> None:
  mergeSortHelp(L, 0, len(L)-1)

def mergeSortHelp(L: list, first: int, last: int)->None:
  if first == last: # conditional statement for base case. 즉 sub의 길이가 1
    return # no action
  else: # recurseve case
    mid = first +(last - first) // 2 # 두 개 sub 로 쪼개기
    mergeSortHelp(L, first, mid) # left sub sort
    mergesortHelp(L, mid+1, last) # right sub sort

    merge(L, first, mid, last) # 두 개 sub의 가장 왼쪽에 있는 두 element들 비교하여, 둘 중 작은 것이 merged list의 가장 왼쪽에 가도록 함

def merge(L: list, first: int, mid: int, last: int) -> None:
  k = first # k는 merged list의 index
  sub1 = L[first:mid+1] # List의 일부를 copy하여 left sub으로
  sub2 = L[mid+1:last+1] # List의 일부를 copy하여 right sub으로
  i = j = 0 # i는 left sub의 index, j는 right sub의 index.

  while i < len(sub1) and j < len(sub2): # left & right sub의 index가 모두 살아있어야 돌아감
    if sub1[i] <= sub2[j]:
      L[k] = sub1[i]
      i = i+1
    else :
      L[k] = sub2[j]
      j = j+1
    k = k+1
  if i < len(sub1): # left sub에 아직 merged list에 가지 않은 element가 있다면, 
    L[k:last+1] = sub1[i:] # 남은 left sub를 merged list 뒤에 그냥 붙여버림.
  elif j < len(sub2):
    L[k:last+1] = sub2[j:]
    
```

#### Time & Memory complexity

 - Merge 
  - Memory Complexity : O(len(L)). 
    - sub left = [first: mid] , sub right = [mid+1, last+1]
  - Time Complexity : O(len(L)). 
    - sub lists are already sorted.

 - MergeSort
  - Memory Complexity: ~N
    - left & right sub에 대한 merge에서, slicing에 의한 메모리 소요
  - Time Complexity: ~ $Nlog_2 N$
    - 절반씩 여러번 쪼개는 알고리즘, 매우 효과적.
    - $N^2$ 에서 $Nlog_2 N$로 줄어든 것
