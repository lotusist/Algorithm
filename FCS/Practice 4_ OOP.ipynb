 - class 생성시 구조에 익숙해져야 한다. 
 - __init__ 함수의 존재 의미: 어떤 object가 특정 class임을 선언할 때, 필요한 변수들을 나열한다. 해당 object는 바로 __init__ 함수에서 지정한 변수들을 지닌다.
 - class의 상속: sub class 생성시, super class의 변수들을 그대로 받을 수 있어 반복해서 코드를 치지 않아도 됨. sub class 생성시, (super class)를 하고, __init__함수에 super().__init__함수를 넣어 상속받을 변수들을 지정.

```python
class Member: 
  def __init__(self, name:str, address: str, email: str, DoB: str, affiliation: str) -> None:
    self.name = name
    self.address = address
    self.email = email
    self.DoB = DoB
    self.affiliation = affiliation
    self.infoList = [self.name, self.address, self.email, self.DoB, self.affiliation]

  def printInfo(self):
    print(self.infoList)
```

```python
class Faculty(Member): 
  def __init__(self, name:str, address: str, email: str, DoB: str, affiliation: str, faculty_num: str) -> None:
    super().__init__(name, address, email, DoB, affiliation)
    self.faculty_num = faculty_num 
    self.advisees = []
    self.courses_teaching = []
    self.infoList += [self.faculty_num, self.advisees, self.courses_teaching]
```

```python
# Cartesian Plane
import math
class Point:
  def __init__(self, x:int, y:int) -> None:
    self.x = x
    self.y = y
```

```python
class Line:
  def __init__(self, start: Point, end: Point) -> None:
    self.start = start
    self.end = end
  def slope(self):
    return (self.end.y - self.start.y) / (self.end.x - self.start.x)
  def length(self):
    return math.sqrt((self.end.y - self.start.y)**2 + (self.end.x - self.start.x)**2)
```

```python
line = Line(Point(1,1), Point(3,2))
```

```python
line.slope()
```

```python
line.length()
```
