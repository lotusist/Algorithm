## 1. Selection Sort

  - 주어진 List의 가장 왼쪽에 가장 작은 element를 놓기
  - 매 iteration마다 List[i:]의 최소 element를 찾아서 L[i]에 위치시켜야 함

```python
def selection_sort(L: list) -> None:
  for i in range(len(L)):
    smallest = find_min(L, i) # the index of the smallest value in L[i:]
    L[i], L[smallest] = L[smallest], L[i]
    
def find_min(L: list, start_idx: int) -> int:
  smallest = start_idx
  for i in range(start_idx+1, len(L)):
    if L[i] < L[smallest]:
      smallest = i
  return smallest
```

```python
def selection_sort(L: list) -> None:
  for i in range(len(L)):
    smallest = i
    for j in range(i+1, len(L)):
      if L[j] < L[smallest]:
        smallest = j
    L[i], L[smallest] = L[smallest], L[i]
```

```python
A = [5,8,7,12,3,8,9,7,-100,54,77]
selection_sort(A)
```

 - Time Complexity
  - N(N+2) for find_min where N is the number of items 

## 2. Insertion Sort
  - 주어진 List의 왼편을 sorted list로 만들기
  - 매 iteration마다 List[i]가 List[:i] 안에서의 적절한 자리를 찾아가야 함

```python
def insertion_sort(L:list) -> None:
  for i in range(1, len(L)):
    insert(L, i) # insert L[i] to the proper location in L[:i]

def insert(L: list, last_idx: int) -> None:
  for i in range(last_idx, 0, -1):
    if L[i-1] > L[i]:
      L[i-1], L[i] = L[i], L[i-1]
    else: 
      break
      
```

```python
def insertion_sort(L:list) -> None:
  for i in range(1, len(L)):
    for j in range(i, 0, -1):
      if L[j-1] > L[j]:
        L[j-1], L[j] = L[j], L[j-1]
      else:
        break
```

 - Time Complexity
  - On average, function insert() may need (i+len(L))/2 for every iteration
  - And i/2 for swapping.
  - N(N+1)/4 - 1/2 for the entire look up, (N-1)N/4 for the entire swap
  - A bit slower than Selection sort
    - find_min() needs to look up the whole list
    - insert() needs to look up only half on average but also need to swap.
    - But when the list is almost sorted, insertion sort needs only kN items for look up.
