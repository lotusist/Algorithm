## 1. Linear Search

 - Ways to implementation
  - While loop
  - While loop and sentinel
  - For loop


```python
def linear_search_while(L:list, value: Any) -> int:
  i = 0
  while i < len(L) and L[i]!=value:
    i = i+1
  if i == len(L):
    return -1
  else:
    return i
```

```python
def linear_search_sentinel(L:list, value:Any) -> int:
  L.append(value)
  i = 0
  while L[i] != value:
    i = i+1
  L.pop()
  if i == Len(L):
    return -1
  else:
    return i
```

```python
def linear_search_for(L:list, value:Any) -> int:
  for i in range(len(L)):
    if L[i] == value:
      return i
  return -1
```

## 2. Binary Search
 - What if the list is sorted?
 - Evaluate the middle of the list and remove the half of the list

```python
def binary_search(L:list, v:Any) -> int:
  start, end = 0
  while start != end:
    mid = (start + end) // 2
    if L[mid] < v:
      start = mid + 1
    else:
      end = mid -1
  
  if start < len(L) and L[start] == v:
    return start
  else:
    return -1
```

 - Time complexity
  - proportional to $\log_{2} len(L)$
  - very effective algorithm
