## Recursion

 - 자기 자신을 계속 부르는 함수
 - If [base case] , else [recursive case] 로 구조화됨
  - base case: 그저 정의에 의해서, 바로 return을 산출할 수 있는 구간
  - recursive case: 이전 단계(서브 그룹)에 대한 function call과 조합하여 현재 단계의 return을 산출하는 구간
 - 검증
  - base case에 대한 return 값이 맞게 출력되는지 확인
  - input n, n-1, n-2 등 2~3개의 case에 대한 return 값이 맞게 출력되는지 확인 (수학적 귀납법)

```python
# Function without recursion
def factorial(n: int) -> int:
  ans = 1
  for i in range(1, n+1):
    ans = ans*i
  return ans
```

```python
factorial(5)
```

```python
# Function with recursion
def factorial_recur(n:int)-> int:
  if n == 0: # conditional statement for base case
    return 1 # base case
  else:
    return n * factorial_recur(n-1) # recursive case
```

```python
factorial_recur(5)
```

```python
# Function for Fibonacci sequence
def Fibonacci(n: int) -> int:
  if n == 1:
    return 1
  elif n == 2:
    return 2
  else :
    return Fibonacci(n-1) + Fibonacci(n-2)
```

```python
def Fibonacci(n: int) -> int:
  if n == 1 or n ==2:
    return n
  else :
    return Fibonacci(n-1) + Fibonacci(n-2)
```
