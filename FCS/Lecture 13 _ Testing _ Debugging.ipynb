## Quality Assurance (QA)
 - 소프트웨어가 제대로 동작하는지 검증하는 단계
 - 코드 최종 단계 전에, 테스트 위한 코드도 만들어야 함. 
 QA 안하면 나중에 더 많은 시간 소모하게 됨
 - 다른 사람들도 읽고 쓸 수 있도록 코드를 잘 작성해야 함. (reusable code)

 - 많은 case를 테스트하는 것보다는, 여러 case를 커버할 수 있는 몇 개의 case만 추리는 것이 필요함.
  - Boundary cases: 여러 조건의 경계에 있는 값들. 해당 case들을 function에 넣으면 오류 매우 빈번함
  - 빈 input
  - 1개 element를 가진 input
  - 2개 element를 가진 input
  - 여러 element를 가진 input
  - 모든 element가 0인 input
  - 모든 element가 양수인 input
  - 모든 element가 음수인 input
  - 양수, 음수, 0 모두를 element로 가진 input
  - 여러 data type을 지닌 input

```python
# Test Example: above_freezing function
def above_freezing(celsius:float) -> bool:
  return celsius > 0
# return celsius >= 0
```

```python
above_freezing(5)
```

```python
above_freezing(-1.25)
```

```python
above_freezing(0)
```

```python
# Test Example: running sum
def running_sum(L:list) -> None: # No return value
  for i in range(len(L)):
    L[i] = L[i-1] + L[i] # i가 0이면 바로 오류.
```

## Guideline for Bug Hunting
 - 내가 뭘 만들고 싶은지 명확하게 알아야 함. 
 - 반복되는 에러를 찾아야 함. 
 - print 등을 활용하여 코드 사이 사이에 넣어 어느 곳이 잘못되었는지를 파악하여야 함.
 - 하나씩 고쳐서 개선해야 함.
 - 디버깅 과정을 반드시 기록하여야 함. 장기적인 시간낭비를 줄여야 함.
