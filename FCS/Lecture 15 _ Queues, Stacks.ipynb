## Queue

 - 선입선출(FIFO) 방식의 자료구조
 - Methods
   - enqueue(): element 하나를 큐에 추가
   - dequeue(): 가장 먼저 큐에 추가된 element를 제거
 - 구현: Linked Lists

## Stack

 - 후입선출(LIFO) 방식의 자료구조
 - Methods
   - push(): element 하나를 스택에 추가
   - pop(): 가장 최근에 들어온 element 하나를 제거
 - 예시: 
   - undo(ctrl+z)
  - 괄호 매칭: 괄호를 열 때마다 스택에 추가. 괄호 닫을 때마다 스택의 최근에 추가된 괄호를 확인
  - C 언어 메모리 저장(할당) 규칙: 가장 나중에 시작한 function이 가장 빨리 끝남
 - 구현: Linked Lists
